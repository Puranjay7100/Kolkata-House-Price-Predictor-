```python
def convert_to_integer(row_value):
  splitted_val = row_value.split(" ")
  amt = float(splitted_val[0])
  if len(splitted_val) == 2:
    unit = splitted_val[1]
    if unit == 'K':
      return 1000 * float(amt)
    elif unit == 'L' or unit == 'Lacs':
      return 100000 * float(amt)
    elif unit == 'Cr':
      return 10000000 * float(amt)
  else:
    return amt
```

```python
df["Flat_Price"]=df["Flat_Price"].apply(convert_to_integer)
df["EMI_Starts"]=df["EMI_Starts"].apply(convert_to_integer)
df['Price_per_sq.ft'] = df['Price_per_sq.ft'].apply(convert_to_integer)
```

```python
X=df.drop(columns="Flat_Price",axis=1)
y=df["Flat_Price"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

XGBoost

```python
from xgboost import XGBRegressor
# Initialize the XGBRegressor
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Fit the model
print(xgb.fit(X_train, y_train))
```

```python
train_pred_xgb = xgb.predict(X_train)
r2_xgb = r2_score(train_pred_xgb,y_train)
r2_xgb
```

```python
test_pred_xgb = xgb.predict(X_test)
r2_xgb = r2_score(test_pred_xgb,y_test)
r2_xgb
```

Prediction System

```python
input_data = (3.0, 'Barrackpore', 1400.0, 3210.0)
input_df = pd.DataFrame([input_data], columns=['BHK', 'Location','Total_Sq.ft','Price_per_sq.ft'])

# Encode the Location
input_df["Location"] = Or.transform(input_df[["Location"]])  # 2D array

# Scale the full input
scaled_input = sc.transform(input_df)

# Predict
prediction = xgb.predict(scaled_input)
print(f"Predicted Price: {prediction[0]:,.2f}")
```
